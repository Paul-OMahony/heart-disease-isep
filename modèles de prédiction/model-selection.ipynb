```python
import pandas as pd
import pickle
```

```python
list_pickles = [
    "df_models_knn.pickle",
    "df_models_rf.pickle",
    "df_models_nb.pickle",
    "df_models_dt.pickle",
    "df_models_svm.pickle"
]

df_summary = pd.DataFrame()

for pickle_ in list_pickles:
    
    path = pickle_
    
    with open(path, 'rb') as data:
        df = pickle.load(data)

    df_summary = df_summary.append(df)

df_summary = df_summary.reset_index().drop('index', axis=1)
```
